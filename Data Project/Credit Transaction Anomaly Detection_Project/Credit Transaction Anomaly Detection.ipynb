## 결제 시간 관련 feature 분석

```python
cc_df['trans_date_trans_time'] = pd.to_datetime(cc_df['trans_date_trans_time'])
```

```python
cc_df['hour'] = cc_df['trans_date_trans_time'].dt.hour
```

```python
cc_df.tail()
```

```python
def hour_func(x):
    if (x >= 6) & (x < 12):
        return 'morning'
    elif (x >= 12) & (x < 18):
        return 'afternoon'
    elif (x >= 18) & (x < 23):
        return 'night'
    else:
        return 'evening'
```

```python
cc_df['hour'].apply(hour_func)
```

```python
cc_df['hour_cat'] = cc_df['hour'].apply(hour_func)
```

```python
cc_df.head()
```

```python
cc_df['hour_cat'].value_counts()
```

```python
all_cnt = cc_df.groupby('cc_num')['amt'].count().reset_index()
```

```python
hour_cnt = cc_df.groupby(['cc_num','hour_cat'])['amt'].count().reset_index()
```

```python
all_cnt.head()
# 인사이트가 어떻게 도출되는지?
```

```python
hour_cnt.head()
```

```python
hour_cnt = hour_cnt.merge(all_cnt, on='cc_num', how='left')
```

```python
hour_cnt.head()
```

```python
hour_cnt = hour_cnt.rename(columns= {'amt_x':'hour_cnt', 'amt_y':'total_cnt'})
# columns= 명시하지 않으면 rename 되지 않음
```

```python
hour_cnt
```

```python
hour_cnt['hour_pct'] = hour_cnt.groupby('hour_cat').apply(
    lambda x: (x['hour_cnt'] / x['total_cnt']) * 100
).reset_index(drop=True)
# 기존 인덱스 버리고 다시 0부터 인덱스 붙임
```

```python
hour_cnt
```

```python
hour_cnt.head(10)
```

```python
hour_cnt.tail(10)
```

```python
cc_df.head()
```

```python
hour_cnt = hour_cnt[['cc_num','hour_cat','hour_pct']]
```

```python
hour_cnt.to_pickle('./hour_cnt.pkl')
```

```python
cc_df = cc_df.merge(hour_cnt, on=['cc_num','hour_cat'], how='left')
```

```python
cc_df.head()
```

```python
cc_df.drop(['trans_date_trans_time', 'hour', 'hour_cat'], axis =1 , inplace = True)
```

## 거리 관련 feature 분석

```python
from geopy.distance import distance
```

```python
cc_df['distance'] = cc_df.apply(lambda x: distance((x['lat'], x['long']), (x['merch_lat'], x['merch_long'])).km, axis = 1)
```

```python
from datetime import datetime
```

```python
start_time =  datetime.now()
cc_df.head(10000).apply(lambda x: distance((x['lat'], x['long']), (x['merch_lat'], x['merch_long'])).km, axis = 1)
datetime.now() - start_time
```

```python
cc_df.head()
```

```python
# Q. cc_num 컬럼의 값마다 distance의 평균, 표준편차를 계산하여 dist_info 변수에 저장합니다.

dist_info = cc_df.groupby('cc_num')['distance'].agg(['mean','std'])
```

```python
dist_info.to_pickle('./dist_info.pkl')
```

```python
# Q. cc_num 컬럼을 기준으로 cc_df와 dist_info 데이터를 합칩니다.
# (cc_df에서 left merge를 수행합니다.)

cc_df = cc_df.merge(dist_info,on='cc_num', how='left')
```

```python
cc_df.head()
```

```python
# Q. distance 값의 z-score를 계산하여 dist_z 컬럼에 저장합니다.

cc_df['distance'].mean()
```

```python
cc_df['distance'].std()
```

```python
cc_df['dist_z'] = cc_df['distance'] - round(cc_df['distance'].mean(),2) / round(cc_df['distance'].std(),2)
```

```python
cc_df
```

```python
# 불필요해진 컬럼들을 제거합니다.

cc_df.drop(['lat','long','merch_lat','merch_long','mean','std'], axis = 1, inplace = True)
```
