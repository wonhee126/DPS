### 승객명: 성과 이름을 분리하여 성 부분만 정리해봅니다.

```python
data['passenger_first_name'] = data['passenger_name'].str.split(expand=True)[0]
# str을 써주어야 오류 발생하지 않음
```

```python
data.head()
```

```python
data.info()
```

```python
# object 형태 -> dateTime 형태
```

```python
data['tpep_pickup_datetime'] = pd.to_datetime(data['tpep_pickup_datetime'])
```

```python
data['tpep_dropoff_datetime'] = pd.to_datetime(data['tpep_dropoff_datetime'])
```

```python
data.info()
```

```python
data['travel_time'] = data['tpep_dropoff_datetime'] - data['tpep_pickup_datetime']
```

```python
data.head()
```

```python
data.info()
```

```python
data['travel_time'] = data['travel_time'].dt.seconds # 모두 초 단위로 변환
```

```python
# int타입으로 변환되었기 때문에 추후에 모델링할 때 오류 발생하지 않는다.
```

```python
# feature engineering
# 승객이 낸 금액 = fare_amount + tip_amount + tolls_amount
# 전체를 더한 값이 별도의 컬럼이 있으면 좋음. total amount로 정의
```

```python
data['total_amount'] = data['fare_amount'] + data['tip_amount'] + data['tolls_amount']
```
