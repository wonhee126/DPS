## Age

-99 가 나온다면 나이차는 -99로 출력  

```python
def age_func(x): # 나이차 구하기
    if x['age'] == -99:
        return -99
    elif x['age_o'] == -99:
        return -99
    elif x['gender'] == 'female':
        return x['age_o'] - x['age'] # 양수면 남자가 더 많음
    else:
        return x['age'] - x['age_o'] # 양수면 여자가 더 많음
        
```

```python
dating_df['age_gap'] = dating_df.apply(age_func,axis=1)
```

만약에 -6과 6 이라는 결과가 나왔다면,  
남자>여자인지 여자>남자인지는 알 수 없지만 두 사람간에 나이 차가 6살 이라는 점은 알 수 있다.  
하지만 컴퓨터는 이를,  
0살(즉, 동갑)보다 나이 차가 덜 난다고 인식하므로 수정해주어야 한다.  

따라서 이 데이터를 '나이 차' 의미로 가져가기 위해서는 절댓값이 필요하다.  
절댓값만 처리한다면, 누가 더 나이 많은지 판단하기 어려우므로  
음수인지 양수인지 판단하여 별도의 컬럼을 만든다.

```python
dating_df['age_gap_dir'] = dating_df['age_gap'].apply(lambda x:'positive' if x>0 else 'negative' if x<0 else 'zero')
```

x>0 이 코드의 의미는, 남자가 연상인 경우다.

```python
dating_df['age_gap'] = abs(dating_df['age_gap'])
```

age_func 함수 해석
반환되는 값이 양수이면 남자가 연상이고,  
음수면 여자가 연상이다.  

age_func 함수를 바탕으로 만들어진 데이터는 age_gap 컬럼에 들어가게 된다.  
age_gap 데이터를 바탕으로 양수면 positive, 음수면 negative, 0이면 zero를 출력한다.  
여기서 positive는 남자가 연상, negative는 여자가 연상, zero는 동갑의 의미다.  
반환된 데이터들을 age_gap_dir 컬럼으로 만든다.

## Rating 평가

o_important컬럼과 o_score컬럼, i_important컬럼과 i_score컬럼을 각각 곱하여 점수를 매긴다.  
o_important 컬럼들을 리스트로 만들고, (각 컬럼도 동일) ... 각각 곱하는 과정은 for문으로 반복한다.  

0보다 큰 값들의 평균을 구해서 o_rating, i_rating 컬럼 생성한다.  
둘의 평균을 구할 때  
100,100과 100,150의 평균이 100인 점을 고려해서 `조합 평균`을 이용한다.  
`조합 평균`의 값은 rating_mean 컬럼으로 저장한다.

    



```python
o_important = []
o_score=[]
i_important = []
i_score = []

for i in dating_df.columns:
    if i.startswith('o_important'):
        o_important.append(i)
    elif i.startswith('o_score'):
        o_score.append(i)
    elif i.startswith('i_important'):
        i_important.append(i) 
    elif i.startswith('i_score'):
        i_score.append(i)
```

```python
dating_df[o_important] = dating_df[o_important].replace({0:-99})
```

```python
dating_df[i_important] = dating_df[i_important].replace({0:-99})
```

```python
def rating(data, important, score):
    if data[score] == -99:
        return -99
    elif data[important] == -99:
        return -99
    else:
        return data[important] * data[score]
```
