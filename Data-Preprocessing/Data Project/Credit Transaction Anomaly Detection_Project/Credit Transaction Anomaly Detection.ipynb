## 결제 시간 관련 feature 분석

사용자별로 자주 사용되는 결제 시간대를 이용  
trans_date_trans_time 컬럼을 이용해서 카드번호별로 평소 결제 시간 외 거래 확인 가능

```python
cc_df['trans_date_trans_time'] = pd.to_datetime(cc_df['trans_date_trans_time'])
```

```python
cc_df['hour'] = cc_df['trans_date_trans_time'].dt.hour
```

```python
cc_df.tail()
```

카드번호별로 어떤 시간대에 얼마만큼 결제가 되었는지 확인해야 한다.  
이는 z-score 로 구할 수 있지만 조금 제약이 있다.  
시간 부분에서, 23숫자와 0숫자는 컴퓨터상에서는 23시간 차지만,  
현실은 1시간 이기 때문에 이 점을 고려해서 숫자가 아니라 
아침, 점심, 저녁, 새벽으로 구분을 한다.

```python
def hour_func(x):
    if (x >= 6) & (x < 12):
        return 'morning'
    elif (x >= 12) & (x < 18):
        return 'afternoon'
    elif (x >= 18) & (x < 23):
        return 'night'
    else:
        return 'evening'
```

```python
cc_df['hour'].apply(hour_func)
```

```python
cc_df['hour_cat'] = cc_df['hour'].apply(hour_func)
```

```python
cc_df.head()
```

```python
cc_df['hour_cat'].value_counts() # 제대로 반영이 됐는지 확인하기 위함
```

```python
# cc_df['distance'] = cc_df.apply(lambda x: distance((x['lat'], x['long']), (x['merch_lat'], x['merch_long'])).km, axis = 1)
cc_df['distance'] = cc_df.apply(lambda x: distance((x['lat'], x['long']), (x['merch_lat'], x['merch_long'])).km, axis = 1)
```

```python
dist_info = cc_df.groupby('cc_num')['distance'].agg(['mean','std']).reset_index()
```

```python
dist_info.to_pickle('./dist_info.pkl')
```

```python
# Q. cc_num 컬럼을 기준으로 cc_df와 dist_info 데이터를 합칩니다.
# (cc_df에서 left merge를 수행합니다.)

cc_df = cc_df.merge(dist_info,on='cc_num', how='left')
```

```python
cc_df.head()
```

```python
# Q. distance 값의 z-score를 계산하여 dist_z 컬럼에 저장합니다.

cc_df['distance'].mean()
```

```python
cc_df['distance'].std()
```

```python
cc_df['dist_z'] = (cc_df['distance'] - round(cc_df['distance'].mean(),2)) / round(cc_df['distance'].std(),2)
```

```python
cc_df
```
